```python
import networkx as nx
import pandas as pd
import os,sys
from IPython.display import display
os.chdir(os.path.expanduser('~/capstone-project'))
sys.path.append('src')

drugs=['streptomycin','sulfamethoxazole','tetracycline','cefalothin','trimethoprim','amoxicillin','ampicillin','levofloxacin','ciprofloxacin']

# make a df of cols Clusters no. and Cluster pairs no. index is drug (capitalized)
df=pd.DataFrame(index=[drug.capitalize() for drug in drugs],columns=['Clusters no.','Cluster pairs no.'])
for drug in drugs:

    G=nx.read_graphml(f'data/temp/Escherichia_coli_{drug}_SVM_0.5.graphml')
    cluster_no=G.number_of_nodes()
    pairs_no=G.number_of_edges()
    df.loc[drug.capitalize()]=cluster_no,pairs_no

df.to_csv('results/SVM_cluster_info.csv')
display(df)
```
